## Proposal

Measuring the World’s Attention — A Global NLP Analysis of Media Coverage During the 2026 Milano-Cortina Winter Olympics

The 2026 Winter Olympics in Milano-Cortina were not just a global sporting event — they were a global media event. But what does “global attention” actually look like in data? When did coverage peak? What narratives dominated? And did different countries frame the Games differently?

In this project, I will use the GDELT global news database (via Google BigQuery) to analyze worldwide media coverage of the 2026 Winter Olympics across time, geography, and language. Instead of relying on viewership reports or anecdotal headlines, I will quantify global attention using large-scale news data.

The analysis will focus on four core questions:
1. **When did media attention peak?**
Using time-series aggregation, I will measure daily coverage intensity before, during, and after the Games — including whether post-Games retrospectives drove more global conversation than the competitions themselves.
2. **How did narratives shift over time?**
I will apply topic modeling (e.g., clustering via embeddings or BERTopic) to headlines and article metadata to identify dominant themes — competition results, broadcast ratings, political reactions, athlete stories, climate narratives, and more — and track how they evolved.
3. **How did sentiment vary across countries and languages?**
Using multilingual transformer models, I will compute sentiment across major source countries (e.g., U.S., Italy, China, U.K.) to examine whether the Games were framed positively, neutrally, or critically in different regions.
4. **What is “real” attention versus syndicated amplification?**
By comparing total article counts, unique headlines, and unique domains, I will separate organic global interest from media syndication effects.

This project combines large-scale data engineering (BigQuery + GDELT), multilingual NLP, and sports analytics to move beyond simple coverage counts and toward a structured understanding of how global media ecosystems react to mega-events.

Ultimately, this article aims to answer a broader question:
When the Games end, how does the narrative begin — and who shapes it?

```python
# Robust GDELT requester

import time
import random
import requests

def gdelt_get_json(url, params, timeout=30, max_retries=8):
    headers = {
        "User-Agent": "Mozilla/5.0 (compatible; SAGB-NLP/1.0; +https://sportsanalytics.studentorg.berkeley.edu/)"
    }
    
    for attempt in range(max_retries):
        r = requests.get(url, params=params, timeout=timeout, headers=headers)

        # Retry on rate-limit + server errors
        if r.status_code in (429, 500, 502, 503, 504):
            wait = min(90, (2 ** attempt) * 2) + random.uniform(0, 1.5)
            print(f"HTTP {r.status_code}. Waiting {wait:.1f}s then retrying ({attempt+1}/{max_retries})...")
            time.sleep(wait)
            continue

        # Other non-200 → raise and show body snippet
        if r.status_code != 200:
            snippet = (r.text or "")[:200].replace("\n", " ")
            raise RuntimeError(f"HTTP {r.status_code}. Body starts with: {snippet}")

        # Validate JSON
        ctype = (r.headers.get("Content-Type") or "").lower()
        if "json" not in ctype:
            snippet = (r.text or "")[:200].replace("\n", " ")
            # sometimes GDELT responds with HTML error pages
            wait = min(90, (2 ** attempt) * 2) + random.uniform(0, 1.5)
            print(f"Non-JSON content-type '{ctype}'. Body starts: {snippet}")
            print(f"Waiting {wait:.1f}s then retrying ({attempt+1}/{max_retries})...")
            time.sleep(wait)
            continue

        try:
            return r.json()
        except Exception as e:
            snippet = (r.text or "")[:200].replace("\n", " ")
            wait = min(90, (2 ** attempt) * 2) + random.uniform(0, 1.5)
            print(f"JSON parse failed: {e}. Body starts: {snippet}")
            print(f"Waiting {wait:.1f}s then retrying ({attempt+1}/{max_retries})...")
            time.sleep(wait)
            continue

    raise RuntimeError("GDELT repeatedly failed (rate limit / non-JSON responses). Try again later or switch to BigQuery.")

def fetch_articles(query, startdatetime, enddatetime, max_total=500, batch=50):
    url = "https://api.gdeltproject.org/api/v2/doc/doc"
    all_rows = []
    startrecord = 1

    while len(all_rows) < max_total:
        params = {
            "query": query,
            "mode": "ArtList",
            "format": "json",
            "sort": "datedesc",
            "startdatetime": startdatetime,
            "enddatetime": enddatetime,
            "maxrecords": batch,
            "startrecord": startrecord
        }

        data = gdelt_get_json(url, params)
        rows = data.get("articles", [])

        if not rows:
            break

        all_rows.extend(rows)
        startrecord += len(rows)

        print(f"Fetched {len(all_rows)} articles...")
        time.sleep(1.0)  # polite delay

    return pd.DataFrame(all_rows)
```

```python
query_tv = '("Milano Cortina" OR "Milan Cortina" OR "Milano-Cortina") AND (Olympics OR Olympic OR "Winter Olympics")'

params = {
    "query": query,
    "mode": "ArtList",
    "format": "json",
    "maxrecords": 10,
    "sort": "datedesc",
    "startdatetime": "20260201000000",
    "enddatetime": "20260228000000",
}


params_tv = {
    "query": query,
    "mode": "timelinevol",
    "format": "json",
    "startdatetime": "20260201000000",
    "enddatetime": "20260228000000",
    "timelinesmooth": 0
}
```

### Insights from the Timeline (Volume Intensity)

- The global conversation about the Olympics may peak not during competition, but during retrospective evaluation.
- Media attention peaks not necessarily during the Games — but during retrospective and politically charged coverage.


*GDELT timelinevol* does NOT return raw article counts, but **normalized intensity** relative to global news coverage. So a spike can mean:
- A real surge in Olympics coverage
- Or a relative surge because overall global news volume dipped
- Or a burst of syndicated/duplicate coverage
